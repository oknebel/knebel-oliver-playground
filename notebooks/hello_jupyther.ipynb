```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data (replace with your own data)
data = {'values': [25, 32, 18, 41, 28, 35, 22, 38, 19, 43]} 
df = pd.DataFrame(data)

# Create histogram
df.hist(column='values', bins=5)  

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Values')

# Show plot
plt.grid(axis='y', alpha=0.5)  # Optional: add gridlines
plt.show()
```
